```python
import pt1_projeto as pt
import PySimpleGUI as sg
import io
from PIL import Image

sg.theme("SandyBeach")

def janelaErro (mensagem):
    interfaceErro=[[sg.Text(mensagem)],[sg.Button("OK")]]
    janelaErro= sg.Window(size=(280,80),title="Mensagem de Erro", font="Candara 13",default_element_size= (15,10)).Layout(interfaceErro)
    stop=False
    while not stop:
        event,values=janelaErro.read()
        if event=="OK" or event==sg.WIN_CLOSED:
            stop=True
    janelaErro.close()

    
def janelaInserir():
    interfaceInserir=[[sg.Text("Introduza os dados do filme")],
                     [sg.Text("Título",size=(10,1)), sg.InputText(key="_TITULO_")],
                     [sg.Text ("Ano",size=(10,1)),sg.InputText(key="_ANO_")],
                     [sg.Text ("Elenco",size=(10,1)),sg.InputText(key="_ELENCO_")],
                     [sg.Text ("Género",size=(10,1)),sg.InputText(key="_GENERO_")],
                     [sg.Button ("Voltar"),sg.Button ("Confirmar")]
                     ]
    janelaInserir=sg.Window("Inserir um Registo",interfaceInserir)
    stop=False
    while not stop:
        event,inputvalue=janelaInserir.read()
        if (event==sg.WIN_CLOSED or event=="Voltar"):
            stop=True
        elif (event=="Confirmar"):
            bd=pt.lerDataset("BDcinema")
            inserir=[]
            pt.inserir(bd,inputvalue["_TITULO_"],inputvalue["_ANO_"],inputvalue["_ELENCO_"],inputvalue["_GENERO_"])
            inserir.append(bd)
            print(inserir)
            janelaPrincipal["_DADOS_"].Update(values=inserir)
        else:
            janelaErro("Erro desconhecido!")
    janelaInserir.close()
#janelaInserir()


def janelaListar(): 
    coluna1=[[sg.Text("Escolha como pretende fazer a listagem.")],
                     [sg.Button ("Género",size=(8,1)),sg.Button("Ordem Alfabética",size=(8,1)),sg.Button("Ator",size=(8,1))],
                    ]
    alfabetica=[]
    coluna2=[[sg.Listbox(values=alfabetica,size=(33,10),key="_alfabetica_")]]
    interfaceListar=[
        [sg.Column(coluna1)],
        [sg.VSeperator()],
        [sg.Column(coluna2)]
    ]
    janela=sg.Window("Listar",font="Candara 22",default_element_size=(20,10)).Layout(interfaceListar) 
    stop=False
    while not stop:
        event,values=janela.read()
        if (event==sg.WIN_CLOSED and event=="Voltar"):
            stop=True
        elif(event=="Ordem Alfabética"):
            bd=pt.carregarDataset("BDcinema")
            alf=pt.listarTitulo(bd)
            i=0
            for i in range (len(alf)):
                alfabetica.append(alf[i])
            if len(alfabetica)==0:
                janelaErro("Não foi carregada nenhuma base de dados!")
            else:
                janela["_alfabetica_"].Update(values=alfabetica)
        
        elif (event=="Género"):
            interfaceGenero=[[sg.Text("Género"),sg.InputText(key="_genre_")],
                            [sg.Button("Voltar"), sg.Button("OK")]]
            janelaGenero=sg.Window("Escolha do género",font="Candara 20").Layout(interfaceGenero)
            stop=False
            while not stop:
                event,inputvalues=janelaGenero.read()        
                if (event==sg.WIN_CLOSED or event=="Voltar"):
                    stop=True
                elif (event=="OK"):
                    bd=pt.carregarDataset("BDcinema")
                    gen=pt.listarGenero(bd,inputvalues["_genre_"])
                    genero=[]
                    i=0
                    for i in range (len(gen)):
                        genero.append(gen[i])
                    if len (genero)==0:
                        janelaErro("Não existe nenhum filme com esse género na base de dados!")
                    else:
                        janela["_alfabetica_"].Update(values=genero)
                        janelaGenero.hide()
                else:
                    janelaErro("Erro desconhecido!")
            janelaGenero.close()
        
        elif (event=="Ator"):
            interfaceAtor=[[sg.Text("Nome do ator"),sg.InputText(key="_ator_")],
                            [sg.Button("Voltar"), sg.Button("OK")]]
            janelaAtor=sg.Window("Escolha do ator",font="Candara 20").Layout(interfaceAtor)
            stop=False
            while not stop:
                event,inputvalues=janelaAtor.read()        
                if (event==sg.WIN_CLOSED or event=="Voltar"):
                    stop=True
                elif (event=="OK"):
                    bd=pt.carregarDataset("BDcinema")
                    at=pt.listarElenco(bd,inputvalues["_ator_"])
                    ator=[]
                    i=0
                    for i in range (len(at)):
                        ator.append(at[i])
                    if len(ator)==0:
                        janelaErro("Não existe nenhum filme desse ator na base de dados!") 
                    else:
                        janela["_alfabetica_"].Update(values=ator)
                        janelaAtor.hide()
                else:
                    janelaErro("Erro desconhecido!")
            janelaAtor.close()
    janela.close()

    
#janelaListar()

def janelaEstatistica():
    interfaceEstatistica=[[sg.Text("Distribuição de Filmes")],
                          [sg.Button("Género",size=(7,1),pad=(20,0)),sg.Button("Ator",size=(7,1),pad=(10,0))],
                          [sg.Button("Voltar",size=(7,1),pad=(65,0))]
                         ]
    janelaEstatistica=sg.Window("Estatística",font="Candara 13",default_element_size=(30,15)).Layout(interfaceEstatistica)
    stop=False
    while not stop:
        bd=pt.lerDataset("BDcinema")
        event,values=janelaEstatistica.read()
        if (event == sg.WIN_CLOSED or event=="Voltar"):
            stop = True
        elif (event=="Género"):
            if len(bd)==0:
                janelaErro("Não foi carregada nenhuma base de dados!")
            else: ###querem ordenar?? como fazer??
                dados=[]
                #pt.plotDistribPorGenero(bd)
                genero=pt.distribPorGenero(bd)
                for c in genero.keys():
                    dados.append(c+"::"+str(genero[c])+"\n")
                janelaPrincipal["_DADOS_"].Update(values=dados)
        elif(event=="Ator"):
            if len(bd)==0:
                janelaErro("Não foi carregada nenhuma base de dados!")
            else:
                dados=[]
                ator=pt.distribPorAtor(bd)
                for c in ator.keys():
                    dados.append(c+"::"+str(ator[c])+"\n")
                janelaPrincipal["_DADOS_"].Update(values=dados)
                
        janelaEstatistica.close()
                
        
                

#janelaEstatistica()

def janelaInformacao():
    janela=[[sg.Text("Informação",font="Candara14")],
            [sg.Text ("-Consultar: deixa ao utilizador pesquisar por um filme através do seu Nome ou ID (não os dois ao mesmo tempo);",font="Candara 12")],
            [sg.Text ("-Carregar: carrega o ficheiro contendo o catálogo de filmes;",font="Candara 12")],
            [sg.Text("-Guardar: grava o atual catálogo de filmes no ficheiro \"BDcinema\;",font="Candara 12")],
            [sg.Text("-Inserir: insere um novo filme no catálogo de filmes, deverá usar um \", \" (vírgula + espaço) para separar os elementos do elenco e os géneros;", font="Candara 12")],
            [sg.Text("-Listar: organiza (por ordem alfabética, género ou ator) e mostra ao utilizador os dados organizados;", font="Candara 12")],
            [sg.Text ("-Dados estatísticos: mostra dados estatísticos, como o número de filmes com um dado género ou o número de filmes em que cada ator participou, através de gráficos;",font="Candara 12")],
            [sg.Text("-Sair: sai da aplicação.",font="Candara 12")],
            [sg.Button("Voltar")]]
        
    janela=sg.Window("Informação",font="Candara 14").Layout(janela)
    stop = False
    while not stop:
        event, values = janela.read()
        if event == sg.WIN_CLOSED or event=="Voltar":
            stop = True
        else:
            janelaErro ("Erro desconhecido!")
    janela.close()

    
#lista onde vai ser carregada a base de dados sempre atualizada.

bd=[]
dados=[]
interfacePrincipal=[[[sg.Text("Catálogo de filmes", font=("Candara 25"), pad=(350,0)), sg.Button ("?",size=(3,1))],
                    [sg.Button ("Consultar",size=(8,1),pad=(10,3)),sg.VSeperator(),sg.Text("Nome"),sg.InputText(key="_NOME_"),sg.VSeperator(),sg.Text("ID"),sg.InputText(key="_ID_")],
                    sg.Column([[sg.Button("Carregar",size=(8,1))],
                    [sg.Button("Guardar",size=(8,1))],
                    [sg.Button("Inserir",size=(8,1))],
                    [sg.Button("Listar",size=(8,1))],
                    [sg.Button("Estatística",size=(8,1))],
                    [sg.Button ("Sair",size=(8,1))]]),
                    sg.Listbox (values=dados,size=(33,10),key="_DADOS_")],
                    [sg.Text ("Trabalho realizado por: Luís Cunha, Miguel Vale, Sara Vieira",font="Candara 13")]
                   ]
janelaPrincipal=sg.Window("Catálogo de Filmes",font="Candara 20",default_element_size=(15,1),resizable=True).Layout(interfacePrincipal)
#event,value=janelaPrincipal.read(close=True)
stop = False 
while not stop:
    bd=pt.lerDataset("BDcinema")
    event, values = janelaPrincipal.read()
    if (event == sg.WIN_CLOSED or event=="Sair" ):
        stop=True
    elif event=="Carregar":
        janelaPrincipal["_DADOS_"].Update(["Foram carregados "+str(pt.numFilmes(bd))+" filmes!"])
    elif (event=="?"):
        janelaInformacao()
    elif (event=="Consultar"):
        i=""
        if (values["_NOME_"]==i and values["_ID_"]==i):
            janelaErro("Erro na inserção dos dados, consulte, por favor, as informações.")
        elif (values["_NOME_"]!=i and values["_ID_"]!=i):
            janelaErro("Erro na inserção dos dados, consulte, por favor, as informações.")
        else:
            dados.append(pt.consultar(bd,values["_ID_"],values["_NOME_"]))
            print (dados)
            if (dados)==0: ##nao funciona, ver melhor!!!
                janelaErro("Não existe nenhum filme com esse nome ou ID! Para adicionar um novo filme prima: Inserir.")
            else:
                janelaPrincipal["_DADOS_"].Update(values=dados)
    #elif(event=="Guardar"):
        
    elif(event=="Inserir"):
         janelaInserir()
    elif (event=="Listar"):
         janelaListar()
    
    elif (event=="Estatística"):
         janelaEstatistica()
    else:
         janelaPrincipal("_DADOS_").update ("Erro:Erro desconhecido!")
janelaPrincipal.close()
```
